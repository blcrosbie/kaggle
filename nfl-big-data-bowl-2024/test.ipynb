```python
def deltas(df):
    final_df = pd.DataFrame()
    df['delta_x'] = 0.0
    df['delta_y'] = 0.0
    gameIds = df.gameId.drop_duplicates().to_list()
    displayNames = df.displayName.drop_duplicates().to_list()
    for gid in gameIds:
        filter_df = df[df['gameId'] == gid]
        for name in displayNames:
            ff_df = filter_df[filter_df['displayName'] == name].reset_index(drop=True)
            
            for rw in ff_df.index:
                try:
                    x_n = ff_df.loc[rw, 'x']
                    x_m = ff_df.loc[rw-1, 'x']
                    delta_x = x_n - x_m
                except:
                    delta_x = 0.0
        
                try:
                    y_n = ff_df.loc[rw, 'y']
                    y_m = ff_df.loc[rw-1, 'y']
                    delta_y = y_n - y_m
                except:
                    delta_y = 0.0
        
                try:
                    t_n = ff_df.loc[rw, 'datetime']
                    t_m = ff_df.loc[rw-1, 'datetime']
                    detla_t = t_n - t_m
                    delta_t = delta_t.total_seconds()
                except:
                    delta_t = 0.0
                    
                ff_df.loc[rw, 'delta_x'] = delta_x
                ff_df.loc[rw, 'delta_y'] = delta_y
                ff_df.loc[rw, 'delta_t'] = delta_t
            
            final_df = pd.concat([final_df, ff_df], axis=0)
    
    return final_df
```

```python
test1_df = pd.read_csv('tracking_week_1.csv')
test1_df['datetime'] = test1_df['time'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S.%f"))
```

```python
new_df = pd.merge(test1_df, test1_df[['pre_index', 'x', 'y', 'datetime', 'gameId', 'playId', 'displayName']], how='left', left_index=True, right_on='pre_index', suffixes=['','_pre'])
```

```python
def calculate_delta(gameId, gameId_pre, playId, playId_pre, name, name_pre, value, value_pre, is_datetime=False):
    if gameId != gameId_pre or name != name_pre or playId != playId_pre:
        return None
    else:
        delta = value - value_pre
        if is_datetime:
            return delta.total_seconds()
        else:
            return delta
        
        
```

```python
new_df['delta_x'] = new_df[['gameId', 'gameId_pre', 'playId', 'playId_pre', 'displayName', 'displayName_pre', 'x', 'x_pre']].apply(lambda x: calculate_delta(gameId=x[0], gameId_pre=x[1], playId=x[2], playId_pre=x[3], name=x[4], name_pre=x[5], value=x[6], value_pre=x[7]), axis=1)
new_df['delta_y'] = new_df[['gameId', 'gameId_pre', 'playId', 'playId_pre', 'displayName', 'displayName_pre', 'y', 'y_pre']].apply(lambda x: calculate_delta(gameId=x[0], gameId_pre=x[1], playId=x[2], playId_pre=x[3], name=x[4], name_pre=x[5], value=x[6], value_pre=x[7]), axis=1)
new_df['delta_t'] = new_df[['gameId', 'gameId_pre', 'playId', 'playId_pre', 'displayName', 'displayName_pre', 'datetime', 'datetime_pre']].apply(lambda x: calculate_delta(gameId=x[0], gameId_pre=x[1], playId=x[2], playId_pre=x[3], name=x[4], name_pre=x[5], value=x[6], value_pre=x[7], is_datetime=True), axis=1)
```

```python
# Clear fields for pre
for col in new_df.columns:
    if 'pre' in col:
        del new_df[col]
```

```python
new_df['delta_x_abs'] = new_df['delta_x'].abs()
new_df['delta_y_abs'] = new_df['delta_y'].abs()
```

```python
new_df = new_df.reset_index(drop=True)
```

```python
new_df = pd.merge(new_df, players_df[['nflId', 'weight', 'height']], on='nflId') 
```

```python
new_df['weight_kg'] = new_df['weight'] * LBS_TO_KG
```

```python
def convert_height(h):
    feet = int(h.split('-')[0])
    inches = int(h.split('-')[1])
    total_inches = feet*12 + inches
    total_yards = round(total_inches/36, 3)
    return total_yards
```

```python
new_df['radius'] = new_df['height'].apply(lambda x: convert_height(x))
```

```python
new_df['mass_kg'] = new_df['weight_kg'] / G
```

```python
new_df['a_ms2'] = new_df['a'] * YDS_TO_M
new_df['dis_m'] = new_df['dis'] * YDS_TO_M
```

```python
new_df['F_horizontal'] = new_df[['mass_kg', 'a_ms2']].apply(lambda x: x[0] * x[1], axis=1)
```

```python
new_df['Work'] = new_df[['F_horizontal', 'dis_m']].apply(lambda x: x[0] * x[1], axis=1)
```

```python
new_df['calories'] = new_df['Work'] * J_TO_cal
```

```python
new_df['Power'] = new_df['Work'] / new_df['delta_t']
```
